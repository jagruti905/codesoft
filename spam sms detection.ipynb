#SPAM SMS DETECTION

#Importing libraries

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
import string
from nltk.corpus import stopwords
from collections import Counter
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('/content/spam.csv', encoding='latin-1')
df.head()
```

```python
df.dropna(how="any", inplace=True, axis=1)
df.columns = ['label', 'message']
```

```python
df['label_num'] = df.label.map({'ham':0, 'spam':1})
df.head()
```

```python
# Remove all the stopword, punctuation
```

```python
import nltk
import string

nltk.download('stopwords')
nltk.download('punkt')

stopwords = nltk.corpus.stopwords.words('english')
punctuations = list(string.punctuation)
stopwords.extend(punctuations)

def clean_msg(msg):
    tokenized_msg = nltk.word_tokenize(msg)
    clean_msg = [word.lower() for word in tokenized_msg if word.lower() not in stopwords]
    return ' '.join(clean_msg)

df['clean_msg'] = df.message.apply(clean_msg)
```

#Model training

```python
X = df.clean_msg
y = df.label_num

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
pipe = Pipeline([('vect', CountVectorizer()),
                 ('tfidf', TfidfTransformer()),
                 ('model', LogisticRegression(solver='liblinear'))])
model = pipe.fit(X_train, y_train)
predictions = model.predict(X_test)
print('Accuracy:', metrics.accuracy_score(y_test, predictions))
```

# Prediction of the msg

```python
def predict_message(message):
  clean_message = clean_msg(message)
  prediction = model.predict([clean_message])
  if prediction[0] == 0:
    return "Ham"
  else:
    return "Spam"

message = "URGENT! You have won a 1 week FREE membership in our �100,000 Prize Jackpot! Txt the word: CLAIM to ."
prediction = predict_message(message)
print(f"Prediction: {prediction}")
```
